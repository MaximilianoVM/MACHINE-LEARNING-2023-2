```python
import random as rand
import numpy as np
import matplotlib.pyplot as plt
from array import *

import pandas as pd
```

```python
rand.seed(5)
```

<h3>generamos los datos</h3>

```python
data_1 = []
data_2 = []

while (len(data_1) + len(data_2)) < 50:
  x = rand.randint(0,400)
  y = rand.randint(0,600)

  #if (x in x_list) == False:
  if np.sqrt(x**2 + y**2) < 350: 
    if ((x,y) in data_1) == False:
      data_1.append((x,y))
      
  elif np.sqrt(x**2 + y**2) > 400:
    if ((x,y) in data_2) == False:
      data_2.append((x,y))
```

estan separados como pares ordenados guardados en las listas data_1 y data_2 para cada conjunto

```python
x_data_1, y_data_1 = zip(*data_1)
x_data_2, y_data_2 = zip(*data_2)
```

visualizamos los conjuntos en graficas distintas

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
axes[0].scatter(x_data_1, y_data_1, color='green', label='data_1')
axes[0].set_xlim([-20, 400])
axes[0].set_ylim([-20, 600])
axes[0].set_title('Data 1')

axes[1].scatter(x_data_2, y_data_2, color='blue', label='data_2')
axes[1].set_xlim([-20, 420])
axes[1].set_ylim([-20, 620])
axes[1].set_title('Data 2')

plt.show()
```

visualizamos todos los datos 

```python
plt.scatter(x_data_1, y_data_1, color='green', label='data_1')
plt.scatter(x_data_2, y_data_2, color='blue', label='data_2')

plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.show()
```

<h1>Discriminante lineal general</h1>

```python
def f(x,  m, b, ret='func', y=''):
  
  if ret == 'func':
    return m*x + b
  
  elif ret == 'eval':
    return y > (m*x + b)
```

<h2> Discriminante 1 </h2>

```python
x_disc = np.linspace(0, 410, 1000)
y_disc = f(x_disc, -1/1.1, 430)

plt.scatter(x_data_1, y_data_1, color='green', label='data_1')
plt.scatter(x_data_2, y_data_2, color='blue', label='data_2')
plt.plot(x_disc, y_disc)

plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.show()
```

```python
def errores(x_data, y_data, m, b):
    arriba = 0
    abajo = 0
    for x, y in zip(x_data, y_data):
        if f(x, m, b, 'eval', y):
            arriba += 1
        else:
            abajo += 1
    return arriba, abajo
```

```python
print(errores(x_data_1, y_data_1, -1/1.1, 430))
```

```python
print('data 1: ', errores(x_data_1, y_data_1, -1/1.1, 430), len(x_data_1))
print('data 2: ', errores(x_data_2, y_data_2, -1/1.1, 430), len(x_data_2))
```
