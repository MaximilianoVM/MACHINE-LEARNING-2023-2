<h1>ECUACION CUADRATICA</h1>

![Alt text](image-4.png)

```python
# Ecuacion cuadratica a resolver: y = ax^2 + bx + c
def quadratic_equation(a, b, c, x):
    return a * x**2 + b * x + c
```

![Alt text](image-7.png)

<h3>fitness function</h3> 

```python
# Funcion de aptitud: cuanto mas cercano a y = 0, mejor
# cuanto mas cumpla ax^2 + bx + c = 0 
def fitness_function(coefs, exp_root):
    a, b, c = coefs
    return abs(quadratic_equation(a, b, c, exp_root))
```

<h3>Funcion de mutacion</h3>
Desplaza los coeficientes

![Alt text](image-8.png)

```python
# Funcion de mutacion: cambia aleatoriamente un valor en el cromosoma
# solution = a, b, c
def mutate(coefs, mutation_rate):
    new_coefs = coefs[:] #copia superficial
    #for i in range(3):
    for coef in new_coefs:
        if random.random() < mutation_rate: #[0.0, 1.0)
            #new_solution[i] += 
            coef += random.uniform(-1, 1) #la mutacion es desplazar los valores entre [0.0, 1.0]
    return new_coefs
```

<h3>Crossover</h3>

![Alt text](image-3.png)

```python
# Funcion de cruce: combina dos soluciones para crear una nueva
def crossover(parent1, parent2):
    index = random.randint(0, 2) #[0, 2]
    child = parent1[:index] + parent2[index:]
    return child
```

<h3>Poblacion inicial</h3> <br>
<i>population: [ [a1, b1, c1], [a2, b2, c2], ... [an, bn, cn] ] <br>
coefs = [a, b, c] </i>


```python
# Crear una poblacion inicial de soluciones aleatorias
def create_population(population_size):
    population = []   
    for _ in range(population_size):
        coefs = [random.uniform(-10, 10) for _ in range(3)] 
        population.append(coefs)
    return population
```

<h2>Algoritmo genético</h2>

```python
# Algoritmo genetico
def genetic_algorithm(population_size, root, n_generations):
    #crea una poblacion inicial
    population = create_population(population_size)
    if population_size < 6 and n_generations < 6:print("population: ", population)
    #10%
    mutation_rate = 0.1
    
    for generation in range(n_generations):
        # Evaluar la aptitud de cada solución en la población
        fitness_scores = [fitness_function(coefs, root) for coefs in population]
        
        # Seleccionar a los mejores individuos
        selected_indices = sorted(range(population_size), key=lambda i: fitness_scores[i]) #ordena de acuerdo al fitness
        selected_population = [population[i] for i in selected_indices[:population_size // 2]] #divide a la mitad a la poblacion
        
        # Crear la nueva generación
        new_population = []
        while len(new_population) < population_size:
            parent1, parent2 = random.choices(selected_population, k=2) # con reemplazo
            child = crossover(parent1, parent2)
            child = mutate(child, mutation_rate)
            new_population.append(child)
        
        population = new_population
        
        #print info 
        if population_size < 6 and n_generations < 6: 
            print('fitness scores:', fitness_scores)
            print("selected_indices", selected_indices)
            print("selected_population", selected_population)
            print("new population: ", population)
            print("\n ============================== \n")
            
            
            
    print('FINAL', population)
    # Devolver la mejor solucion encontrada
    best_solution = population[0]
    return best_solution
```
